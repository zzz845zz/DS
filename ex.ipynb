```python
def get_category(df, category):
    is_valid = df[category] != 0
    return df[is_valid]

def split_testAndValidate(df, category):
    column_people = [x for x in df.columns if x[0]=='남' or x[0]=='여']
    column_category = [x for x in df.columns if x not in column_people and x!='행정동']
    
    is_train = df['연월'] < 202000
    is_test = df['연월'] >= 202000
    
    x_train = np.array(df[is_train][column_people])
    y_train = np.array(df[is_train][category])
    
    x_test = np.array(df[is_test][column_people])
    y_test = np.array(df[is_test][category])
    
    #return df[column_people], df[column_category]
    return x_train, y_train, x_test, y_test
```

```python
data_tmp1 = get_category(data, '음식료품')
data_tmp1.shape
# data_tmp1_train = data_tmp1[data_tmp1['연월'] < 202000]
# data_tmp1_train
```

```python
x_train, y_train, x_test, y_test = split_testAndValidate(data_tmp1, '음식료품')
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(normalize=True)
model.fit(X=x_train, y=y_train)
lr_pred = model.predict(X=x_test)
lr_pred.shape
```

## performance

```python
# Root Mean Square Error
# numpy
print('numpy RMSE', np.sqrt(np.mean(np.square((lr_pred.flatten()-y_test.flatten())))))      
# scikit-learn
from sklearn import metrics
print('scikit-learn RMSE', np.sqrt(metrics.mean_squared_error(lr_pred, y_test)))      

# R2
print('R2', metrics.r2_score(y_test.flatten(), lr_pred.flatten()))
```

```python
plt.scatter(x=y_test, y=lr_pred)
plt.title('Linear Regression')
print(y_test.shape, lr_pred.shape)
print(y_test.flatten().shape, lr_pred.flatten().shape)
#print(np.correlate(y_test.flatten(), lr_pred.flatten()))
#print(np.corrcoef([y_test[:,0], lr_pred[:, 0]]))
```
